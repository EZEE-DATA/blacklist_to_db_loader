```python
from pydantic import BaseModel

class AuthorityItem(BaseModel):
    id: int
    name: str

class BlockedItem(BaseModel):
    authority: AuthorityItem
    domains: list[str]
    id: int
    ips: list[str]
    status: str
    urls: list[str]

class SnapshotBlocked(BaseModel):
    item: BlockedItem
    last_updated: str | None
```

```python
%%time
from uploader import MongoCollection
from settings_loader import SettingsFile

token_filename = 'tokens/mongo.token'
cache_filename = 'cache/ips'
db_name = 'black_lists'
collection_name = 'ips'

with open(cache_filename, 'r') as cache_file:
    cache_values = list(map(lambda x:x.replace('\n',''),cache_file.readlines()))

db_string = SettingsFile(token_filename).settings
```

```python
%%time
#get existing_ips
with MongoCollection(db_string=db_string,
                     db_name=db_name,
                     collection_name=collection_name) as collection:
    existing_records = collection.find({'ip':{'$in':cache_values}})
    existing_ips = list(map(lambda x: x['ip'],list(existing_records)))
# get new ips
cache_values_set = set(cache_values)
existing_ips_set = set(existing_ips)
new_ips_set = cache_values_set - existing_ips_set
new_ips = list(new_ips_set)
print(f'exists:{len(existing_ips)}, new:{len(new_ips)}')

with MongoCollection(db_string=db_string,
                     db_name=db_name,
                     collection_name=collection_name) as collection:
    if len(existing_ips) != 0:
        collection.update_many({'ip':{'$in':existing_ips}}, {"$currentDate":{"last_update":True}})
    if len(new_ips) != 0:
        collection.insert_many([{"ip":ip} for ip in new_ips])
        collection.update_many({'ip':{'$in':new_ips}}, {"$currentDate":{"last_update":True}})
```

```python
print(len(cache_values))
with open(cache_filename, 'r') as cache_file:
    prim_cache_values = list(map(lambda x:x.replace('\n',''),cache_file.readlines()))
print(len(prim_cache_values))
print(len(prim_cache_values)-len(cache_values))
```
